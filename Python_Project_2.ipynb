```python
# importing the dataset
import pandas as pd
import matplotlib.pyplot as plot
df = pd.read_csv('https://raw.githubusercontent.com/niteen11/DataAnalyticsAcademy/master/Python/dataset_diabetes/diabetic_data.csv',)
df
```

```python
# Examining more of the data
df.head(50)
```

```python
# renaming the columns to my liking
df.rename(columns={'patient_nbr': 'Patient Number', 'time_in_hospital': 'Days in Hospital'}, inplace=True)
df
```

```python
# Checking to see if I have any null values
df.isnull().sum()
```

```python
# Looking at the datatypes of the columns
df.dtypes
```

```python
# Dropping the '?'s in columns 'weight' and 'race' from my dataframe because its incomplete data
df.drop(df.index[df['weight'] == '?'], inplace=True)
df.drop(df.index[df['race'] == '?'], inplace=True)
```

```python
df
```

```python
# Checking to see what other columns have '?'
(df=='?').sum()
```

```python
# Dropping the rows with '?' in column 'Payer Code'
df.drop(df.index[df['payer_code'] == '?'], inplace=True)
```

```python
# Checking again to see if any more columns have rows with '?' data
(df=='?').sum()
```

```python
# Dropping rows with '?' in 'medical_specialty'
df.drop(df.index[df['medical_specialty'] == '?'], inplace=True)
```

```python
# Checking again to see if '?'s remain in the dataframe
(df=='?').sum()
```

```python
# Dropping the '?'s from column 'diag_3'
df.drop(df.index[df['diag_3'] == '?'], inplace=True)
```

```python
# Checking again to see if I still have any '?'s in the dataframe
(df=='?').sum()
# finally down to 0
```

```python
#Looking to see what my dataframe looks like now
df
```

```python
# Pulling statistics of the current dataframe
df.describe()
```

```python
# Analyzing the average days spent in the hospital by race
df.groupby(['race'])['Days in Hospital'].mean().sort_values(ascending=False).plot(kind="bar")
# Days spent in the hospital were highest in those that are African American and Caucasian
# Both races on average spent about 4 days in the hospital
# On average, Other spent about 3 days in the hospital
# On average, Asians spent the least amount time in the hospital, just 1 day
```

```python
# Analyzing the median days spent in the hospital by race
df.groupby(['race'])['Days in Hospital'].median().sort_values(ascending=False).plot(kind="bar")
# These results nearly perfectly mirror the mean results above.
```

```python
# Analyzing the average number of days spent in the hospital by gender
df.groupby(['gender'])['Days in Hospital'].mean().sort_values(ascending=False).plot(kind="bar")
# On average, females spent slightly more days in the hospital than men did
```

```python
# Analyzing the median number of days spent in the hospital by gender
df.groupby(['gender'])['Days in Hospital'].median().sort_values(ascending=False).plot(kind="bar")
# When analyzing this metric, both genders are on par with each other for days spent in the hospital
```

```python
# Analyzing the average number of days spent in the hospital by age
df.groupby(['age'])['Days in Hospital'].mean().sort_values(ascending=False).plot(kind="bar")
# Surprisingly, this data is not ordered from oldest to youngest group
# People aged 80-90 stayed for 5 days on average
# People aged 20-80 stayed for 4 days on average
# People aged younger than 20 stayed in the hospital for about 2 days on average
```

```python
# Analyzing the median number of days spent in the hospital by age
df.groupby(['age'])['Days in Hospital'].median().sort_values(ascending=False).plot(kind="bar")
# The median chart also has a surprising result that does not align from oldest to youngest
# While 80-90 is the highest at 5 days, 90-100 ranks in the middle at 4 days
# The median value for 30-80 year olds is 4 days
# Excluding 30-40 year olds, people under 50 years old stayed in a hospital for median time of 3 days or 2 days in the case of 10-20 year olds
```

```python
# Analyzing the average number of days spent in the hospital by weight
df.groupby(['weight'])['Days in Hospital'].mean().sort_values(ascending=False).plot(kind="bar")
# Once again there does not appear to be a trend for number of days in hospital based on weight
# The two heaviest weights led to the longest hospital stays, but the average across the categories only varies from 4 to 5
```

```python
# Analyzing the median number of days spent in the hospital by weight
df.groupby(['weight'])['Days in Hospital'].median().sort_values(ascending=False).plot(kind="bar")
# This chart is closer to being aligned to the highest weight resulting in the longest stay while the lowest weight resulting in the shortest stay
```

```python
# Analyzing the average number of days spent in the hospital by medical specialty 
df.groupby(['medical_specialty'])['Days in Hospital'].mean().sort_values(ascending=False).plot(kind="bar")
# Denistry resulted in the highest stay in hospital. At nearly 6 days, this was the longest stay of any category across these charts
# Pediatrics resulted in one of the lowest stays in hospital. This aligns with previous charts which indicated young people had shorter hospital stays.
```

```python
# Analyzing the median number of days spent in the hospital by medical specialty 
df.groupby(['medical_specialty'])['Days in Hospital'].median().sort_values(ascending=False).plot(kind="bar")
# The median chart is similar to the mean chart.
# In this chart, pediatric does move to the lowest duration of hospital stay
```
